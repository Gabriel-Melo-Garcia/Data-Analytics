```python
# 8. Extract the rows where the employee has been with the company for more than 5 years and has a Performance Score of at least 4.
from datetime import datetime

df['Joining Date'] = pd.to_datetime(df['Joining Date'])
current_date = datetime.now()
df['years_in_company'] = round((current_date - df['Joining Date']).dt.days / 365.25 , 1 )

yearsInCompany_5_score_4 = df[(df['years_in_company']>5) & (df['Performance Score']>=4)]
yearsInCompany_5_score_4
```

```python
# 10.Retrieve rows where the employee belongs to "Sales" and has a Salary in the top 10% of all salaries.
# mede after ask chatgpt if my code was the best way to do it.
top_10_percent_salary = df['Salary'].quantile(0.9)
select_top_10_percent = df.loc[(df['Department'] == 'Sales') & (df['Salary'] >= top_10_percent_salary)]
select_top_10_percent 
```

```python
# 11.Replace all "Inactive" values in the Status column with "Retired".

df['Status'] = df['Status'].replace('Inactive', 'Retired')
df
```

```python
# 12.Add a new column Bonus where values are 10% of the Salary column.
df['Bonus'] = df['Salary'] * 0.1
df
```

```python
# 13.Update the Performance Score of employees in the "IT" department to 5 if it is NaN.
df.loc[(df['Department'] == 'IT') & (df['Performance Score'].isna()), 'Performance Score'] = 5.0
df[df['Department']== 'IT']
```

```python
# 14.Modify the Experience of employees with less than 2 years to 2.
df.loc[(df['Experience'] < 2),'Experience'] = 2.0
df
```

```python
# 15.Normalize the Salary column so that all values are between 0 and 1.
df['Normalized Salary'] = (df['Salary'] - df['Salary'].min())/(df['Salary'].max()-df['Salary'].min())
df
```

```python
# 16.Replace all Location values with "Remote" for employees with a Performance Score below 3.
df.loc[(df['Performance Score'] < 3 ) | (df['Performance Score'].isna()), 'Location'] = 'Remote'
df
```

```python
#17.Create a new column Experience Level with values "Junior", "Mid", or "Senior" based on the Experience.
q1 = df['Experience'].quantile(0.25)
q3 = df['Experience'].quantile(0.75)

def categorize_experience(experience):
    if experience <= q1:
        return "Junior"
    elif q1 < experience <= q3:
        return "Mid"
    else:
        return "Senior"
df['Experience Category'] = df['Experience'].apply(categorize_experience)

df[['Experience', 'Experience Category']]
```

```python
#18.Add a new row with specific values using the append() method.
new_employee = {
    'ID': (df['ID'].max()+1),
    'Name': 'Gabriel Garcia',
    'Age': 23,
    'Gender': 'Male',
    'Department': 'IT',
    'Salary': 2500,
    'Joining Date': datetime(2024, 10, 12),  
    'Performance Score': 3.0,
    'Experience': 2,
    'Status': 'Active',
    'Location': 'Dublin',
    'Session': 'Morning',
    'years_in_company': round((current_date - datetime(2024, 10, 12)).days / 365.25,1) ,
    'Bonus': 2500*0.1
    
}


new_row = pd.DataFrame([new_employee])
df = pd.concat([df, new_row], ignore_index=True)
df
```

```python
# 19.Delete the column Session from the DataFrame.
df_whitout_session = df.drop('Session',axis=1)
df_whitout_session
```

```python
# 20.Delete rows where the Age is greater than 60.
select_age_over_60 = df[df['Age']>60].index
df.drop(select_age_over_60, axis=0, inplace=True)
df
```

```python
# 21.Add a new column Age Group with values "Young", "Middle-aged", or "Senior" based on Age..

def age_categoryze(age):
    if age <= 25 :
        return 'Young'
    elif 25 < age <= 50 :
        return 'Middle-aged'
    else:
        return 'Senior'
df['Age Group'] = df['Age'].apply(age_categoryze)

df
```

```python
# 48.Create a DataFrame with employee ID and Project Count, then merge it with the original dataset. Calculate the average Salary per project.
df_project =  pd.DataFrame({
    "ID": range(1, 1001),  
    "Project Count": np.random.randint(1, 6, size=1000)  
})
print(df_project.head())

df_merged_4 = pd.merge(df, df_project, on = 'ID', how = 'left')

df_merged_4["Salary per Project"] = df_merged_4["Salary"] / df_merged_4["Project Count"]
df_average_salary_per_project = df_merged_4["Salary per Project"].mean()

print('\nmean salary per project',round(df_average_salary_per_project,2))
```
